```python
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
age = pd.read_csv('age_brackets.csv')
age.head()
```

```python
education = pd.read_csv('education_brackets.csv')
education.head()
```

```python
income = pd.read_csv('household_income.csv')
income.head()
```

```python
brackets = pd.merge(age, education,
                   left_on = ['STATE','PLACE','YEAR','GEOID'], right_on = ['STATE','PLACE','YEAR','GEOID'],
                   how = 'inner')
brackets.head()
```

```python
brackets = pd.merge(brackets, income,
                   left_on = ['STATE','PLACE','GISJOIN','YEAR','GEOID'], right_on = ['STATE','PLACE','GISJOIN','YEAR','GEOID'],
                   how = 'inner')
brackets.head()
```

```python
brackets['percent_hs_diploma'] = brackets.hs_diploma / brackets.Total
brackets['percent_some_college'] = brackets.some_college / brackets.Total
brackets['percent_college_degree'] = brackets.college_degree / brackets.Total
brackets['percent_higher_degree'] = brackets. higher_degree / brackets.Total
```

```python
bracket = brackets[['GISJOIN','YEAR','STATE','PLACE','GEOID','Total','total_under_25','total_25_44','total_45_64',
            'total_65_up','percent_under_25','percent_25_44','percent_45_64','percent_65_over','hs_diploma',
            'some_college','college_degree','higher_degree','percent_hs_diploma','percent_some_college',
            'percent_college_degree','percent_higher_degree','median_income','error_margin']]
```

```python
bracket.tail()
```

```python
bracket.to_csv('bracket_captsone.csv')
```

```python
bracket['college'] = bracket.college_degree + bracket.higher_degree
bracket.head()
```

```python
bracket['percent_total_college_degree']= bracket.college / bracket.Total 
bracket.head(1)
```

```python
bracket['college_rank'] = bracket['college'].rank()
bracket.head()
```

```python
bracket['income_rank'] = bracket['median_income'].rank(ascending = False)
bracket['income_rank_perc'] = bracket['median_income'].rank(pct = True)
bracket.head()
```

```python
bracket['rank_under_25'] = bracket['total_under_25'].rank()
bracket['rank_25_44'] = bracket['total_25_44'].rank()
bracket['rank_45_64'] = bracket['total_45_64'].rank()
bracket['rank_65_up'] = bracket['total_65_up'].rank()
```

```python
bracket.to_csv('bracket2_capstone.csv')
```
