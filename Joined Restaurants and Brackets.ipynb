## Subtracting needed columns from Franklin's numbers, ranking, and pulling out 20 

```python
fixed_brackets['pop_franklin'] = 83096.0 - fixed_brackets['Total']
fixed_brackets['under_25_franklin'] = 0.323722 - fixed_brackets['percent_under_25']
fixed_brackets['25_44_franklin'] = 0.30395 - fixed_brackets['percent_25_44']
fixed_brackets['45_64_franklin'] = 0.249759 - fixed_brackets['percent_45_64']
fixed_brackets['65_up_franklin'] = 0.122569 - fixed_brackets['percent_65_over']
fixed_brackets['median_income_franklin'] = 103813 - fixed_brackets['median_income']
fixed_brackets['college_franklin'] = 0.478898 - fixed_brackets['percent_total_college_degree']
```

```python
f = fixed_brackets.sort_values('median_income_franklin')
```

```python
f['income_rank_franklin'] = f['median_income_franklin'].rank()
f.head(20)
```

```python
f = f[(f.income_rank_franklin >= 1) & (f.income_rank_franklin <= 20)]
f.shape
```

```python
f.to_csv('income_rank_franklin.csv')
```
